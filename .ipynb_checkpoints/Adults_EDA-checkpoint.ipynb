### Histogramas de las variables

```python
cols = df_train.columns
```

```python
# Valores que contiene cada atributo
for feature in cols:
    print('feature values: ', feature, df_train[feature].unique())
```

```python
# Contando los valores nulos en el conjunto de datos
for columna in cols:
    conteo = df_train[columna].replace(['?', ' ', ' ?'], np.nan).isnull().sum()
    if conteo > 0:
        porcentaje = (conteo /  df_train['age'].shape[0]) * 100 
        print(f"Cantidad de valores nulos en la columna {columna}: {conteo} = {porcentaje.round(2)}%")
```

No existen valores nulos en el conjunto de datos de entrenamiento

Se realizara una transformacion al atributo income que sera nuestro atributo a predecir, para los valores <50k, su valor sera 0 y para >50k su valor sera 1

```python
num_cols = df_train.select_dtypes(include=['int', 'float']).columns.tolist()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(16, 8))

def create_histogram(ax, data, column_name, x_label, y_label):
    ax.hist(data)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# Loop through each subplot and create a histogram
for i, name in enumerate(num_cols):
    row = int(i / 3)
    col = i % 3
    create_histogram(axs[row, col], df_train[name], name, name, 'Frequency')

#fig.tight_layout()
fig.suptitle('Histogramas de los atributos de Adults Income')
plt.show()
fig.savefig("./Imagenes/adult-hist.eps", transparent=True, bbox_inches = 'tight') 
```

```python
df_train['income'] = df_train['income'].replace({'<=50K': 0,  '>50K': 1})
```

Los histogrmas muestran que los atributos capital gain y capital loss tienen una cantidad superior de valores de 0, con respecto a valores mayores a 0, por lo que se usara el agrupamiento de dos valores,
0 cuando el valor es 0 y 1 cuando es mayor a 0.

Se comprobara la correlacion del atributo fnlwgt para ver si aporta informacion para la construccion del modelo de clasificacion

```python
correlation_matrix = df_train[num_cols].corr()

sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

El atributo fnlwgt tiene una correlacion negativa muy cercana a 0, por lo que se tomara la decision de eliminar este atributo debido a que no aporta mucha informacion para la construccion del modelo de prediccion.

## Preprocesamiento

### Limpieza de los datos

```python
df_temp = df_train.copy()
```

```python
# Eliminacion de las columnas
df_train = df_train.drop('education', axis = 1)

df_train = df_train.drop('fnlwgt', axis = 1)

# Eliminacion de las columnas de prueba
df_test = df_test.drop('education', axis = 1)

df_test = df_test.drop('fnlwgt', axis = 1)
```

### Transformacion de los datos

```python
cols_transform = ['workclass', 'native.country', 'sex', 'capital.gain', 'capital.loss']
```

```python
le = LabelEncoder()

print('sex')
print(df_train['sex'].unique())
df_train['sex'].unique()
df_train['sex'] = le.fit_transform(df_train['sex'])
df_train['sex'].unique()
print(df_train['sex'].unique())

#Prueba
print('sex - prueba')
print(df_test['sex'].unique())
df_test['sex'].unique()
df_test['sex'] = le.transform(df_test['sex'])
df_test['sex'].unique()
print(df_test['sex'].unique())
```

```python
# Atributo capital-gain
# Se puede reducir a dos valores (=0 y >0 ) del rango [0, 1]
df_train['capital.gain'].value_counts()
df_train['capital.gain'] = np.where(df_train['capital.gain'] > 0, '>0', '0')
df_train['capital.gain'].unique()

# Atributo capital-loss
# Se puede reducir a dos valores (=0 y >0 ) del rango [0, 1]
df_train['capital.loss'].value_counts()
df_train['capital.loss'] = np.where(df_train['capital.loss'] > 0, '>0', '0')
df_train['capital.loss'].unique()

# Atributo capital-gain prueba
# Se puede reducir a dos valores (=0 y >0 ) del rango [0, 1]
df_test['capital_gain'].value_counts()
df_test['capital_gain'] = np.where(df_test['capital_gain'] > 0, '>0', '0')
df_test['capital_gain'].unique()

# Atributo capital-loss prueba
# Se puede reducir a dos valores (=0 y >0 ) del rango [0, 1]
df_test['capital_loss'].value_counts()
df_test['capital_loss'] = np.where(df_test['capital_loss'] > 0, '>0', '0')
df_test['capital_loss'].unique()
```

```python
print('capital.gain')
print(df_train['capital.gain'].unique())
df_train['capital.gain'].unique()
df_train['capital.gain'] = le.fit_transform(df_train['capital.gain'])
df_train['capital.gain'].unique()
print(df_train['capital.gain'].unique())

#Prueba
print('capital_gain - Prueba')
print(df_test['capital_gain'].unique())
df_test['capital_gain'].unique()
df_test['capital_gain'] = le.transform(df_test['capital_gain'])
df_test['capital_gain'].unique()
print(df_test['capital_gain'].unique())
```

```python
print('capital.loss')
print(df_train['capital.loss'].unique())
df_train['capital.loss'].unique()
df_train['capital.loss'] = le.fit_transform(df_train['capital.loss'])
df_train['capital.loss'].unique()
print(df_train['capital.loss'].unique())

#Prueba
print('capital_loss - Prueba')
print(df_test['capital_loss'].unique())
df_test['capital_loss'].unique()
df_test['capital_loss'] = le.transform(df_test['capital_loss'])
df_test['capital_loss'].unique()
print(df_test['capital_loss'].unique())
```

```python
# Atributo marital-status
df_train['marital.status'] = df_train['marital.status'].replace(['Married-civ-spouse', 'Married-AF-spouse'], 'married')
df_train['marital.status'] = df_train['marital.status'].replace(['Never-married', ' Divorced', 'Married-spouse-absent', 'Separated', 'Widowed'], 'not-married')
df_train['marital.status'].unique()

# Atributo marital-status Prueba
df_test['marital_status'] = df_test['marital_status'].replace(['Married-civ-spouse', 'Married-AF-spouse'], 'married')
df_test['marital_status'] = df_test['marital_status'].replace(['Never-married', ' Divorced', 'Married-spouse-absent', 'Separated', 'Widowed'], 'not-married')
df_test['marital_status'].unique()
```

```python
df_train['marital.status'] = (df_train['marital.status'] == 'married').astype(int)

# Prueba
df_test['marital_status'] = (df_test['marital_status'] == 'married').astype(int)
```

```python
# Atributo Native Country
df_train['native.country'] = (df_train['native.country'] == 'United-States').astype(int)

# Atributo Native Country Prueba
df_test['native_country'] = (df_test['native_country'] == 'United-States').astype(int)
```

```python
df_train['race'] = (df_train['race'] == 'White').astype(int)

# Prueba
df_test['race'] = (df_test['race'] == 'White').astype(int)
```

```python
df_train['workclass'] = (df_train['workclass'] == 'Private').astype(int)

# Prueba
df_test['workclass'] = (df_test['workclass'] == 'Private').astype(int)
```

```python
# Observando los valores del conjunto de datos de entrenamiento
# Valores que contiene cada atributo
cols = df_train.columns
for feature in cols:
    print('feature values: ', feature, df_train[feature].unique())
```

```python
# Observando los valores del conjunto de datos de prueba
# Valores que contiene cada atributo
cols = df_test.columns
for feature in cols:
    print('feature values: ', feature, df_test[feature].unique())
```

```python
# Utilizando one hot encoding
cols_ohe = ['occupation', 'relationship']

#Encoder
encoder = OneHotEncoder(sparse_output = False).set_output(transform = 'pandas')
```

```python
df_train_transform = encoder.fit_transform(df_train[cols_ohe])
#df_train_transform

# Prueba
df_test_transform = encoder.transform(df_test[cols_ohe])
#df_test_transform
```

```python
df_train_t = pd.concat([df_train, df_train_transform], axis = 1).drop(columns=cols_ohe)

# Prueba
df_test_t = pd.concat([df_test, df_test_transform], axis = 1).drop(columns=cols_ohe)
```
