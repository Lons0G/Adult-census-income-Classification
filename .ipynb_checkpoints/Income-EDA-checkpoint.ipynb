```python
# Valores que contiene cada atributo
cols = df_test.columns
for feature in cols:
    print('feature values: ', feature, df_test[feature].unique())
```

```python
# Contando los valores nulos en el conjunto de datos
for columna in cols:
    conteo = df_train[columna].replace(['?', ' ', ' ?'], np.nan).isnull().sum()
    if conteo > 0:
        porcentaje = (conteo /  df_train['age'].shape[0]) * 100 
        print(f"Cantidad de valores nulos en la columna {columna}: {conteo} = {porcentaje.round(2)}%")
```

No existen valores nulos en el conjunto de datos de entrenamiento

Se realizara una transformacion al atributo income que sera nuestro atributo a predecir, para los valores <50k, su valor sera 0 y para >50k su valor sera 1

```python
num_cols = df_train.select_dtypes(include=['int', 'float']).columns.tolist()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(16, 8))

def create_histogram(ax, data, column_name, x_label, y_label):
    ax.hist(data)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# Loop through each subplot and create a histogram
for i, name in enumerate(num_cols):
    row = int(i / 3)
    col = i % 3
    create_histogram(axs[row, col], df_train[name], name, name, 'Frequency')

#fig.tight_layout()
fig.suptitle('Histogramas de los atributos de Adults Income')
plt.show()
#fig.savefig("./Imagenes/adult-hist.eps", transparent=True, bbox_inches = 'tight') 
```

Los histogrmas muestran que los atributos capital gain y capital loss tienen una cantidad superior de valores de 0, con respecto a valores mayores a 0, por lo que se usara el agrupamiento de dos valores,
0 cuando el valor es 0 y 1 cuando es mayor a 0.

Se comprobara la correlacion del atributo fnlwgt para ver si aporta informacion para la construccion del modelo de clasificacion

### Preprocesamiento

```python
df_train = df_train.drop('education', axis = 1)
```

```python
df_train = pd.concat([df_train.drop('occupation', axis =  1), pd.get_dummies(df_train['occupation']).add_prefix('occ_')], axis = 1)
df_train = pd.concat([df_train.drop('workclass', axis =  1), pd.get_dummies(df_train['workclass']).add_prefix('work_')], axis = 1)
df_train = pd.concat([df_train.drop('marital.status', axis =  1), pd.get_dummies(df_train['marital.status']).add_prefix('marital_')], axis = 1)
df_train = pd.concat([df_train.drop('relationship', axis =  1), pd.get_dummies(df_train['relationship']).add_prefix('rel_')], axis = 1)
df_train = pd.concat([df_train.drop('race', axis =  1), pd.get_dummies(df_train['race']).add_prefix('race_')], axis = 1)
df_train = pd.concat([df_train.drop('native.country', axis =  1), pd.get_dummies(df_train['native.country']).add_prefix('nc_')], axis = 1)
```

```python
df_train['sex'] = df_train['sex'].apply(lambda x: 1 if x == 'Male' else 0)
df_train['income'] = df_train['income'].apply(lambda x: 1 if x == '>50K' else 0)
```

```python
df_train = df_train.rename(columns={'education.num': 'education_num'})
df_train = df_train.rename(columns={'capital.gain': 'capital_gain'})
df_train = df_train.rename(columns={'capital.loss': 'capital_loss'})
df_train = df_train.rename(columns={'hours.per.week': 'hours_per_week'})
```
